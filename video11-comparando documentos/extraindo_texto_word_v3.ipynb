4. Criando uma função que varre cada paragrafo de um arquivo word

```python
def varre_word(nome_arquivo):
    docx_dict={}
    lista_paragrafos=[]
    documento=Document(nome_arquivo)
    for para in documento.paragraphs:
        if(len(para.text)>0):
            docx_dict=(para.text,nome_arquivo)
            lista_paragrafos.append(docx_dict)
    return lista_paragrafos
```

```python
doc=varre_word('teste.docx')
doc
```

5.Mapeando todos os arquivos word em uma pasta

```python
lista_arquivo=[]
for arquivo in glob.glob(r"C:\Users\renat\OneDrive\Documentos\Renata\Estudo\Trabalho Inteligente\VÍDEOS\video10 - extract_word\arquivos\*.docx" ):
    lista_arquivo.append(arquivo)
```

6. Varrendo todos os arquivos

```python
tabelas=[]
lista_word=[]
for arquivo in lista_arquivo:
    lista_word=varre_word(arquivo)
    tabelas.append(pd.DataFrame(lista_word))
    lista_word=[]
```

```python
tabelas[1]
```

```python
#consolidando tudo em uma única tabela
tabela_final=pd.concat(tabelas)
tabela_final
```

```python
#Organizando as informações 
tabela_final.columns=['Paragrafo','Nome do Arquivo']
tabela_final.reset_index(drop=True, inplace=True)
tabela_final
```

```python
#tratando o nome do arquivo
tabela_final['Nome do Arquivo']=tabela_final['Nome do Arquivo'].str.replace(r"C:\\Users\\renat\\OneDrive\\Documentos\\Renata\\Estudo\\Trabalho Inteligente\\VÍDEOS\\video10 - extract_word\\arquivos\\", "")
tabela_final
```

7.Transformando em dados estruturados

```python
tabela_final_enconded=pd.get_dummies(tabela_final['Paragrafo'])
tabela_final_enconded=tabela_final.join(tabela_final_enconded)
tabela_final_enconded
```

```python
tabela_final_enconded=tabela_final_enconded.drop(['Paragrafo'],axis=1)
tabela_final_enconded
```

```python
tabela_consolidada=pd.pivot_table(tabela_final_enconded,index="Nome do Arquivo", aggfunc=np.sum)
tabela_consolidada
```

8. Exportando para excel

```python
tabela_consolidada.to_excel('tabela.xlsx')
```
