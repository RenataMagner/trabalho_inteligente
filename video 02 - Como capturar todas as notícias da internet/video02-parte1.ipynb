```python
import requests as rq
import json 
import pandas as pd 
```

```python
url='http://web.archive.org/cdx/search/cdx?url=economia.uol.com.br//&collapse=digest&from=20201026&to=20201126&output=json'
urls=rq.get(url).text
urls
```

```python
parse_url=json.loads(urls)
parse_url
```

```python
parse_url[1]
```

```python
parse_url[1][1]
```

```python
waylink=parse_url[1][1]+'/'+parse_url[1][2]
waylink
```

```python
final_url='https://web.archive.org/web/'+waylink
final_url
```

```python
url_list=[]
for bloco in range(1,len(parse_url)):
    url_original=parse_url[bloco][2]
    timestamp=parse_url[bloco][1]
    waylink=timestamp+'/'+url_original
    url_list.append(waylink)
```

```python
url_list
```

```python
list_url_final=[]
for url in url_list:
    list_url_final.append('https://web.archive.org/web/'+url)
```
